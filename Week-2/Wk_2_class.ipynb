# Reading modes

```python
with open('data.txt', 'r') as file:
    content = file.read()
    print(content)

# use read function to read from a file (read) )
```

# Writing Modes

```python
with open('data.txt','w') as file:
    file.write('freeman\n')
    file.write('Nelson\n')
    file.write('Victory\n')

    #\n line break
    # Nelson
    # freeman
    # Victoria
```

```python
# Writing mode can override all the data in your file if you want to add a new data, so it;s not advisable to use write keyword or mode, but you can use append to add it
```

# Append Mode

```python
# In a situation where you have a file that already have a data, and you don't want to erase anything, this is where apend mode will help to 
# add more data to it

# for instance we already have this:

with open('data.txt','w') as file:
    file.write('freeman\n')
    file.write('Nelson\n')
    file.write('Victory\n')
```

```python
To append mode, then add below: 
```

```python
# this returns: # Nelson
    # freeman
    # Victoria
    # Obi is a man
    # Ade is a great man
```

```python
# Apend mode is used to add additional data to the already given data, don;t ever use write mode b/c it overrides all the data
# In reading mode, any attempt to add or write data will always result in an error, vis-a-vis writing mode.
```

# Writing csv file

```python
with open('crypto_data.csv','w', newline ='') as file:
    writer = csv.writer(file)

    writer.writerow(['token_name', 'price', 'category'])
    writer.writerow(['BTC', '67000', 'cex'])
    writer.writerow(['ETH', '2000', 'dex'])
    writer.writerow(['BNB', '1800', 'dex'])
```

# updating data from json file

```python
# All the item in json must be in a single enclosed bracket. How dictionary work. first step is to Read data from json, then write it back
```

```python
token_data = {
    'token' : 'BTC',
    'price' : 2000,
    'category' : 'Cex',
    'volume' : 90000
}
# the above is the innefficient way of updating it


# efficiant way of updating it

token_data['volume'] = 90000
print(token_data)
```

```python
# So the 3 steps for updating the json file is:
# first we read the data
# second we update the data
# third we write back the data

# for example below
    
```

```python
# Reading the data:

with open('token_detail.json', 'r') as file:
    data = json.load(file)
    print(data)


# Updating the data

token_data['volume'] = 90000

# Thirdly, write this data back into our system

with open('token_detail.json', 'w') as file:
    data = json.dump(data,file,indent=4)

# This will add volume to the above on vs code like this:

# {'token': 'BTC', 'price': 2000, 'category': 'Cex', 'volume': 90000}
```

```python
| Function       | Purpose                           |
| -------------- | --------------------------------- |
| `json.load()`  | Read JSON from a file             |
| `json.dump()`  | Write JSON to a file              |
| `json.loads()` | Read JSON from a string           |
| `json.dumps()` | Convert dictionary to JSON string |
```
