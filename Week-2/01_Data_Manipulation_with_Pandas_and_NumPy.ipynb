# Data Manipulation with Pandas and NumPy

Data manipulation is the process of changing, organizing, or transforming data to make it more useful, readable, or suitable for analysis. It involves tasks like cleaning, filtering, sorting, grouping, or calculating new values from existing data.

Tools like Pandas and NumPy, popular Python libraries, are often used to streamline these tasks. Pandas excels at handling structured data, like tables of crypto trades, for filtering or grouping, while NumPy supports fast numerical computations, such as calculating average prices or returns.

For instance, a crypto investor might manipulate data by filtering trades to show only Bitcoin transactions over $5,000, sorting them by timestamp to track price movements, and aggregating daily totals to assess trading volume. This helps uncover trends, optimize strategies, and ensure data accuracy, ultimately supporting better decisions in the fast-paced crypto market.





## Pandas 

Pandas is a popular Python library used for working with data. It helps you load, clean, analyze, and manipulate data easily. Think of it like an Excel spreadsheet in Python! It’s great for handling tables of data.

### Key Concepts
Series: A single column of data (like a list with labels).

Index: Labels for rows, helping you identify and access data.

You can load data from files (CSV, Excel, etc.), manipulate it, and save it back.





### Installing/Importing Pandas

```python
# In Commmand Line: Enter pip install pandas

import pandas as pd
```

### Create a DataFrame
A DataFrame is like a table with rows and columns. Let’s create one with sample cryptocurrency data (e.g., coin names, prices, and trade volumes).



```python
# Create a dictionary with crypto data
data = {
    'Coin': ['Bitcoin', 'Ethereum', 'Ripple', 'Litecoin'],
    'Price': [45000, 3000, 0.85, 120],
    'Volume': [15000, 8000, 2000000, 50000]
}

# Turn it into a DataFrame
df = pd.DataFrame(data)

# View the DataFrame
print("Original DataFrame:")
print(df)
```

### Explore the Data

Let’s check out the DataFrame to understand it better.<br>
View the first few rows: Use head() to see the top 5 rows (or fewer if less data).



```python
print("\nFirst 2 rows:")
print(df.head(2))
```

### Get basic info: 
Check column names, data types, and missing values with info().


```python
print("\nDataFrame Info:")
df.info()
```

###  Basic Data Manipulation
Now, let’s manipulate the data with common tasks.<br>
Select Columns<br>
Access one or more columns easily.



```python
# select a specific column

print("\nSelecting 'Coin' and 'Price' columns:")
print(df[['Coin', 'Price']])
```

### Filter Rows
Keep rows based on a condition, e.g., Price > 100.



```python
# Filter rows 

print("\nFiltering rows where Price > 100:")
print(df[df['Price'] > 100])
```

### Sort Data
Sort by a column, like Price, in ascending or descending order.



```python
# Sorting data

# Sorting by Price in ascending order
print("\nSorted by Price (ascending):")
print(df.sort_values('Price'))

# Sorting by Price in descending order
print("\nSorted by Price (descending):")
print(df.sort_values('Price', ascending=False))
```

### Add a New Column
Calculate a new value, e.g., a hypothetical 2% price increase.



```python
# Adding a new column with a calculation

df['Price_After_2Pct'] = df['Price'] * 1.02
print("\nDataFrame with new 'Price_After_2Pct' column:")
print(df)
```

### Save and Load Data

You can save your DataFrame to a file and load it later.<br>
Save to CSV:



```python
# saving the DataFrame to a CSV file

df.to_csv('crypto_data.csv', index=False)
print("\nData saved to 'crypto_data.csv'")
```

## Numpy

NumPy is a powerful Python library for numerical computations, especially useful for working with arrays (lists of numbers) and performing fast mathematical operations. It’s a foundation for data manipulation, often used alongside Pandas, and is great for tasks like calculations, statistics, and handling crypto-related data. Below, I’ll walk you through the basics step by step—installing NumPy, creating arrays, and performing common operations with simple code. This is beginner-friendly and straightforward.



### Install And Importing NumPy
First, ensure NumPy is installed. Run this in your terminal or command prompt:



```python
# Run This command On your Command Prompt:  pip install numpy

# Importing numpy
import numpy as np
```

###  Create a NumPy Array
A NumPy array is like a list but optimized for math and faster operations. Let’s create arrays with sample cryptocurrency data (e.g., coin prices and trade volumes).



```python
# Create arrays for crypto data
prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin
volumes = np.array([15000, 8000, 2000000, 50000])  # Trade volumes

# View the arrays
print("Prices array:")
print(prices)
print("\nVolumes array:")
print(volumes)
```

### Explore the Array
Let’s check basic properties of the arrays.
Shape: See the structure (number of elements).



```python
# checking the shape of the arrays

prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin
print("\nShape of prices array:")
print(prices.shape)
```

### Data Type: Check the type of data in the array.



```python
# chceking the data type of the arrays

prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin
print("\nData type of prices array:")
print(prices.dtype)
```

###  Basic Data Manipulation
NumPy excels at math and array operations. Let’s try common tasks.<br>
    Basic Math Operations<br>
    Perform calculations on all elements at once.



```python
# Add 100 to all prices

prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin
prices_plus_100 = prices + 100
print("\nPrices after adding 100:")
print(prices_plus_100)
```

```python
# Calculate a Percentage Increase
# Compute a 2% price increase.

prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin
price_increase = prices * 1.02
print("\nPrices after 2% increase:")
print(price_increase)
```

```python
# Filter Data
# Select elements based on a condition, e.g., prices > 100.

prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin
high_prices = prices[prices > 100]
print("\nPrices greater than 100:")
print(high_prices)

```

### Basic Statistics
NumPy makes stats easy. Let’s compute some for our arrays.



```python
# Calculate statistics for prices

prices = np.array([45000, 3000, 0.85, 120])  # Prices for Bitcoin, Ethereum, Ripple, Litecoin

print("\nStatistics for Prices:")

print("Mean (Average):", np.mean(prices)) # Average price

print("Minimum:", np.min(prices)) # Minimum price

print("Maximum:", np.max(prices)) # Maximum price

print("Sum:", np.sum(prices)) # Total sum of prices

print("Standard Deviation:", np.std(prices)) # Standard deviation of prices
```

### Work with 2D Arrays
NumPy can handle multi-dimensional arrays. Let’s combine our data into a 2D array (like a table).



```python
# Create a 2D array (rows: coins, columns: price and volume)

crypto_data = np.array([
    [45000, 15000],
    [3000, 8000],
    [0.85, 2000000],
    [120, 50000]
])

print("\n2D Crypto Data Array (Price, Volume):")
print(crypto_data)

# Access a row (e.g., Ethereum, index 1)
print("\nEthereum data (row 1):")
print(crypto_data[1])

# Access a column (e.g., all prices, column 0)
print("\nAll prices (column 0):")
print(crypto_data[:, 0])
```

### Save and Load Data
Save your array to a file and load it back.<br>
Save to file:



```python
# Save the 2D array to a .npy file

crypto_data = np.array([
    [45000, 15000],
    [3000, 8000],
    [0.85, 2000000],
    [120, 50000]
])
np.save('crypto_array.npy', crypto_data)
print("\nData saved to 'crypto_array.npy'")
```

#### Load from file:


```python
# Lading the saved .npy file

loaded_data = np.load('crypto_array.npy')
print("\nLoaded 2D array:")
print(loaded_data)
```
