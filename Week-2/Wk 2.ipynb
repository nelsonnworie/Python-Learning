# Reading modes

```python
with open('data.txt', 'r') as file:
    content = file.read()
    print(content)

# use read function to read from a file (read) )
```

# Writing Modes

```python
with open('data.txt','w') as file:
    file.write('freeman\n')
    file.write('Nelson\n')
    file.write('Victory\n')

    #\n line break
    # Nelson
    # freeman
    # Victoria
```

```python
# Writing mode can override all the data in your file if you want to add a new data, so it;s not advisable to use write keyword or mode, but you can use append to add it
```

```python
Append Mode
```

```python
plt.hist(students, scores, color = 'red')

plt.title('class performance')

plt.xlable('student name')

plt.ylabel('student scores')

plt.savefig('hist.png')
```
