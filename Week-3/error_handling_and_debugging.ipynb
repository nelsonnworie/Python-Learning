## 3. Exception /Error Handling

Python provides tools to handle runtime errors using **exceptions**.

Exception handling allows programs to:

- Prevent crashes
- Display meaningful error messages
- Continue execution safely

Python handles exceptions using the **try and except blocks**.

### Try and Except

The `try` block contains code that might produce an error.

The `except` block contains code that runs if an error occurs.

Structure:

**try**: Run Code

**except**: Handle error


```python
# try:
    # Search for Kangarooo Pictures
# except:
    # Print("These are results for kangaroo pictures Search instead for kangarooo pictures")
```

```python
try:
    number = int(input("Enter a number: "))
    print(number)
except:
    print("An error occurred. Please try again.")
```

```python
try:
    int(input("Enter a number: "))
except:
    print("Invalid input. Please enter a valid integer.")
```

```python
try:
    number = int(input("Enter a number: "))
    print(number)
except:
    print("Enter a valid number")
```

```python
try:
    print(10/"Hello")
except:
    print("An error occurred. Please try again.")
```

### Catching Specific Exceptions

It is best practice to catch **specific exceptions** instead of catching all errors.

This improves debugging and prevents hiding unexpected issues.

```python
try:
    int("Hello")
except ValueError:
    print("Passwords must be numbers")
```

### Handling Multiple Exceptions

Python allows multiple exceptions to be handled separately.

This provides more precise error handling.

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("That wasn't a number.")
except ZeroDivisionError:
    print("You cannot divide by 0, enter another number")
```

### The Else Block

The `else` block runs **only if no exception occurs**.

This helps separate normal execution from error handling.

Think of it like ordering food online.

try → Place the order  
except → If the payment fails, show an error  
else → If the payment succeeds, prepare and deliver the food

```python
try:
    num = int(input("Enter number: "))
except ValueError:
    print("Invalid number")
else:
    print("You entered:", num)
```

### The Finally Block

The `finally` block always runs whether an error occurs or not.

It is typically used for **cleanup operations** such as closing files or releasing resources.

```python
try:
    open("data.txt")
except FileNotFoundError:
    print("File not found")
finally:
    print("Enter a valid number next time") 
```

## Debugging

Debugging is the process of:

1. Identifying bugs
2. Understanding why they occur
3. Fixing them

Developers commonly use techniques such as:

- print debugging
- logging

### Print Debugging

The simplest debugging technique is printing variable values to observe program behavior.


```python
x = 10
y = 5

result = x * y

print(result)

# result = x / y

# print("result =", result)
```

```python
# Debugging with logging

import logging
logging.basicConfig(level=logging.DEBUG)
logging.debug("This is a debug message.")
```

## Best Practices for Error Handling

Professional developers follow these principles:

1. Catch specific exceptions
2. Avoid hiding errors
3. Keep try blocks small
4. Use logging for production systems
5. Always read tracebacks carefully
