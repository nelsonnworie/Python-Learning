# Introduction to APIs (RESTful API Requests with Python)

Modern applications often need to communicate with other systems to exchange data.  
One of the most common ways this communication happens is through **APIs (Application Programming Interfaces)**.

An **API** is a set of rules and protocols that allows different software applications to communicate with each other. 

An API allows different software systems to exchange data.

## Examples
- Weather apps fetching live data  
- Apps retrieving user profiles  
- Websites processing payments  

## Key Idea
APIs act as a **bridge** between systems without exposing internal logic.

# What is a REST API?

**REST (Representational State Transfer)** is a standard way APIs communicate over the internet using HTTP.

## How it works
1. Client sends a request  
2. Server processes it  
3. Server returns data (usually JSON)  

## Analogy
You go to a restaurant (Client), a Waiter takes your order (Request) to the kitchen (Server) and brings back food (Data)

- Client → You  
- Request → Order  
- Server → Kitchen  
- Response → Food  

## Note
The server sends a **representation** (JSON), not raw database data.

# API Request Components (Important)

When making API requests, you often use:

---

## 1. URL (Endpoint)
The base address of the API.

Example:
https://api.coingecko.com/api/v3/ping

---

## 2. Params (Query Parameters)

Used to send **filters or options** in the URL.

Example URL:
https://api.com/data?user=1

Python:
params = {
    "user": 1
}

requests.get(url, params=params)

### When to use:
- Filtering data
- Searching
- Pagination

---

## 3. Headers

Headers send **extra information about the request**.

Example:
headers = {
    "Content-Type": "application/json",
    "Authorization": "Bearer API_KEY"
}

### Common Uses:
- Authentication (API keys)
- Telling server the data format
- Custom metadata

---

## 4. Body (Request Payload)

Used in POST/PUT requests to send data.

Example:
data = {
    "name": "Alice"
}

requests.post(url, json=data)

### Important:
- GET → no body
- POST/PUT → body required

---

## 5. Response

The server sends back:
- Data (JSON)
- Status code

Example:
response.json()
response.status_code

## POST Requests

A **POST request** sends data to the server to create or update resources.

Example use cases:

- Creating a user account
- Submitting form data
- Uploading information to a database

````python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

data = {
    "title": "Learning APIs",
    "body": "This is a POST request",
    "userId": 1
}

response = requests.post(url, json=data)

print(response.status_code)
print(response.json())
````

## HTTP Response Status Codes

When an API request is made, the server returns a **status code**.

Common status codes include:

| Code | Meaning |
|-----|------|
| 200 | Success |
| 201 | Resource created |
| 400 | Bad request |
| 404 | Resource not found |
| 500 | Server error |

## Handling Errors

API requests can fail due to:

- Network issues
- Invalid endpoints
- Server errors

Python allows error handling using if-else blocks.

````python
# Error Handling

import requests

url = "https://api.coingecko.com/api/v3/ping"

response = requests.get(url)

if response.status_code == 200:
    print(response.json())
else:
    print("Request failed:", response.status_code)
````

Write a script that:

````python
import requests
# import json

url = "https://api.coingecko.com/api/v3/simple/price"

params = {
    "ids": "bitcoin,ethereum",
    "vs_currencies": "usd"
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    print("Bitcoin Price is", data["bitcoin"]["usd"])
    print("Ethereum Price is", data["ethereum"]["usd"])
else:
    print("Error fetching data")

````
