```python
# Python provides tools to handle runtime errors using exceptions.

Exception handling allows programs to:

. Prevent crashes
. Display meaningful error messages
. Continue execution safely

Python handles exceptions using the try and except blocks.

Try and Except

The try block contains code that might produce an error.

The except block contains code that runs if an error occurs.

Structure:

try: Run Code

except: Handle error
```

```python
try:
    number = int(input("Enter a number: "))
    print('number')
except:
    print("an error occured, please try again. ")
```

```python
try:
    result = 10/0
except ZeroDivisionError:
    print("You cannot divide by zero")
```

```python
try:
    int("Hello")
except ValueError:
    print("Password must be numbers")
```

```python
Handling Multiple Exceptions

Python allows multiple exceptions to be handled separately.

This provides more precise error handling.
```

```python
try:
    num = int(input ("Enter a number: ") )
    result = 10 / num
except ValueError:
    print("That wasn't a number.")
except ZeroDivisionError:
    print("You cannot divide by 0, enter another number")
```

```python
v The Else Block

The else block runs only if no exception occurs.

This helps separate normal execution from error handling.

Think of it like ordering food online.

try - Place the order
except - If the payment fails, show an error
else If the payment succeeds, prepare and deliver the food
```

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid number")
else:
    print("You entered:", num)
```

```python
The Finally Block

The finally block always runs whether an error occurs or not.

It is typically used for cleanup operations such as closing files or releasing resources.

file = open("data.txt")
```

```python
try:
    # Something that might raise FileNotFoundError
    with open("myfile.txt") as f:
        data = f.read()
except FileNotFoundError:
    print("File not found")
finally:
    print("Program finished")
      
```

Best Practices for Error Handling

Professional developers follow these principles:

1. Catch specific exceptions
2. Avoid hiding errors
3. Keep try blocks small
4. Use logging for production systems
5. Always read tracebacks carefully

```python
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero!"
    else:
        return result

# Example usage:
print(safe_divide(10, 2))  # Should print 5.0
print(safe_divide(10, 0))  # Should print Error: Cannot divide by zero!
```

```python
try:
    with open("example.txt", "r") as f:
        content = f.read()
        print("File content:")
        print(content)
except FileNotFoundError:
    print("File not found! Make sure 'example.txt' exists.")
finally:
    print("Program finished.")
```

What is a REST API?

REST (Representational State Transfer) is a standard way APis communicate over the internet using HTTP.

How it works

1. Client sends a request
2. Server processes it
3. Server returns data (usually JSON)

Analogy
You go to a restaurant (Client), a Waiter takes your order (Request) to the kitchen (Server) and brings back food (Data)
. Client -. You
. Request -. Order
. Server .. Kitchen
. Response - Food

Note

The server sends a representation (JSON), not raw database data.

1. Endpoint (URL)
