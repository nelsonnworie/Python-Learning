```python
from web3 import Web3
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()
INFURA= os.getenv("INFURA_URL")

print(f'infura import as {INFURA}')
```

```python
w3 =  Web3(Web3.HTTPProvider(INFURA))
print(f'Is the connection successful?: {w3.is_connected()}')
```

```python
def turn_to_checksum(address: str) -> str:
    if not Web3.is_checksum_address(address):
        return Web3.to_checksum_address(address)
    return address


ACCOUNT_ADDRESS_RAW = "0x28C6c06298d514Db089934071355E5743bf21d60"
ACCOUNT_ADDRESS = turn_to_checksum(ACCOUNT_ADDRESS_RAW)
print(f'Is the address valid?: {Web3.is_checksum_address(ACCOUNT_ADDRESS)}')
```

```python
# 1.Pick up an address and get the balance, using .get_balance method. Put this get balance inside a function

# 1. Get Balance
def get_bal(addr):
    return w3.from_wei(w3.eth.get_balance(addr), 'ether')

NEW_ADDR = "0x28C6c06298d514Db089934071355E5743bf21d60"
print(get_bal(NEW_ADDR))
```

```python
# Task 2: Filter transactions by address

block = w3.eth.get_block('latest', True)
txs = [t for t in block.transactions if t['from'] == NEW_ADDR or t['to'] == NEW_ADDR]

print(f"2. Filtered Transactions Result: {len(txs)}")
```

```python
# 3.  Get latest block and details using the latest method.
latest = w3.eth.get_block('latest')
print(latest['number'], latest['hash'].hex())
```
