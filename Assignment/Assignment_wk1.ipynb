```python
# Question 2   Conditionals and Logical Operators
# Topic: Control Structures: if, elif, else, Logical Operators
# Write a program that does the following:
# Ask the user to enter a number.
# Use if, elif and else to check whether the number is positive, negative, or zero.
# Use logical operators (and, or, not) to check whether the number is between 10 and 100 inclusive, and whether it is even or odd.
# Print a clear summary of all three checks based on the number entered.
```

```python
num = int(input("Enter a number: "))

# Check positive, negative, or zero
if num > 0:
    pos_neg_zero = "positive"
elif num < 0:
    pos_neg_zero = "negative"
else:
    pos_neg_zero = "zero"

# Using 'and' explicitly
if num >= 10 and num <= 100:
    between_10_100 = "between 10 and 100"
else:
    between_10_100 = "not between 10 and 100"

# Using 'not' explicitly
if not num % 2 == 0:
    even_odd = "odd"
else:
    even_odd = "even"

# Print summary
print(f"The number is {pos_neg_zero}.")
print(f"It is {between_10_100}.")
print(f"It is {even_odd}.")
```

```python
# Question 3   Loops
# Topic: Control Structures: for loop, while loop, break, continue
# Write a program with the following two parts:
# Part A: Create a list of at least 5 items (for example, a shopping list, fruit names, or country names). Use a for loop to print each item in the list along with its position number, starting from 1. The output should look like: 1. Apple, 2. Mango, and so on.
# Part B: Write a while loop that keeps asking the user to enter a password. If the password entered matches a password you have set in your code, print Access granted and stop the loop. If it does not match, print Wrong password, try again and let the user try again. After 3 failed attempts, print Too many attempts. Access denied. and exit the loop using break.
```

```python
# Part A: Print list items with position numbers

fruits = ["Apple", "Mango", "Banana", "Orange", "Pineapple"]

for i in range(len(fruits)):
    print(i + 1, ".", fruits[i])
```

```python
# Part B: Password checking with while loop

password = "python123"   # password set in the code
attempts = 0             # count number of tries

while attempts < 3:
    user_input = input("Enter password: ")

    if user_input == password:
        print("Access granted")
        break
    else:
        print("Wrong password, try again")
        attempts += 1

if attempts == 3:
    print("Too many attempts. Access denied.")
```

```python
# Part B: Password checking with while loop

password = "python123"   # password set in the code
attempts = 0             # count number of tries

while attempts < 3:
    user_input = input("Enter password: ")

    if user_input == password:
        print("Access granted")
        break
    else:
        print("Wrong password, try again")
        attempts += 1

if attempts == 3:
    print("Too many attempts. Access denied.")
```

```python
# Function to greet a user
def greet(name):
    print("Hello", name, ", welcome to Python class.")

# Call the function
greet("Nelson")
```

```python
# Function to calculate BMI
def calculate_bmi(weight, height=1.75):
    bmi = weight / (height ** 2)
    print("BMI:", bmi)
    return bmi
```

```python
calculate_bmi(70, 1.75)
```

```python
calculate_bmi(70)
```

```python
# Function to calculate total and average score
def total_score(*args):
    total = sum(args)
    average = total / len(args)

    print("Total:", total)
    print("Average:", average)

    return total, average
```

```python
total_score(80, 90, 70)
```

```python
total_score(60, 75, 85, 90)
```

```python
total_score(100, 95)
```

```python
# Question 6   Working with CSV Files
# Topic: File Handling: CSV files using the csv module
# Write a program that does the following:
# Create a CSV file called products.csv with the columns Product, Price and Quantity. Add at least 4 rows of product data to the file using csv.writer().
# Read the file back using csv.DictReader() and print each row to the screen.
```

```python
import csv

# Create CSV file and write data
with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)

    # Write header
    writer.writerow(["Product", "Price", "Quantity"])

    # Write rows
    writer.writerow(["Laptop", 1200, 5])
    writer.writerow(["Mouse", 25, 20])
    writer.writerow(["Keyboard", 45, 15])
    writer.writerow(["Monitor", 300, 7])
```

```python
# Read CSV file and print rows
with open("products.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print(row)
```

```python
import pandas as pd

# Load CSV file
df = pd.read_csv("btc-history.csv")

# Explore the DataFrame
print("Columns:", df.columns.tolist())
print("\nFirst 5 rows:\n", df.head())
print("\nShape of the DataFrame:", df.shape)
print("\nBasic statistics:\n", df.describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.plot(df['snapped_at'][:50], df['price'][:50], color='blue')  # first 50 rows
plt.title("Bitcoin Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("line_chart.png")
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['snapped_at'][:10], df['price'][:10], color='green')  # first 10 rows
plt.title("Bitcoin Price Bar Chart (First 10 Days)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("bar_chart.png")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(df['price'][:5], labels=df['snapped_at'][:5], autopct='%1.1f%%')
plt.title("Price Distribution for First 5 Days")
plt.savefig("pie_chart.png")
plt.show()
```
