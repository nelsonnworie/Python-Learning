Build a Python script that fetches, processes, and displays the top 100 cryptocurrency market data using the CoinGecko API.

```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",      # price in USD
    "order": "market_cap_desc", # top coins by market cap
    "per_page": 100,            # return 100 coins
    "page": 1,                  # first page
    "sparkline": False          # no sparkline data, keeps it clean
}

response = requests.get(url, params=params)
data = response.json()

for coin in data:
    print(f"Rank {coin['market_cap_rank']} | {coin['name']} ({coin['symbol'].upper()}) | ${coin['current_price']:,}")
```

```python
import pandas as pd

df = pd.DataFrame(data)

table = df[["market_cap_rank", "name", "current_price"]]
table.columns = ["Rank", "Coin", "Price"]

table.head(100)
```
