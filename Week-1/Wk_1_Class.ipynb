# conditional statement

```python
# if
age = 10

if age == 10:
    print(f'The age is (age)')
    
```

```python
# else

age = 10

if age == 10:
    print(age)
else:
    print('the age is', age)
```

```python
#elif

age = 10

if age == 1:
    print('if statement executed!')
elif age == 10:
    print('elif statement executed!')
```

```python
# if, else, elif

name = 'Ade'

if name == 'Chike':
    print('The name is Chike')
elif name == 'Obi':
    print('The name is Obi')
elif name == 'Bisi':
    print("The name is Bisi")
else:
    print(name)
```

```python
#Logical operation 

age = 10
height = 2.5

if age == 10 and height == 2.4:
    print('This is correct')
```

```python
# or 
age = 10
height = 2.5

if age == 10 or height == 2.5:
    print('This is corrct')
```
